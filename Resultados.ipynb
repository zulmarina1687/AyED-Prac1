```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_tam, array_cpp = np.loadtxt('BScpp.txt', delimiter=',', unpack=True)
array_tam, array_py = np.loadtxt('BSpy.txt', delimiter=',', unpack=True)
array_tam, array_go = np.loadtxt('BSgo.txt', delimiter=',', unpack=True)

df = pd.DataFrame({'Tamaño':array_tam,
                   'C++':array_cpp,
                   'Python':array_py,
                   'Goland':array_go})
fig, ax = plt.subplots()
df.plot(x = 'Tamaño', y = 'C++', ax = ax, marker = 'o')
df.plot(x = 'Tamaño', y = 'Python', ax = ax, marker = 'o')
df.plot(x = 'Tamaño', y = 'Goland', ax = ax, marker = 'o')
ax.set_title('Algoritmo BubbleSort', loc = "center", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_xlabel("Tamaño", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Time ", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.grid(axis = 'y', color = 'gray', linestyle = 'dashed')
plt.savefig("BubbleSort")
plt.show()
```
